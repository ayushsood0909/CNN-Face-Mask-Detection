```python
from keras.models import load_model
import cv2
import numpy as np
```

```python
model = load_model('model-010.model')

face_clsfr=cv2.CascadeClassifier(r"D:\Downloads\haar-cascade-files-master\haar-cascade-files-master\haarcascade_frontalface_default.xml")

source=cv2.VideoCapture(2)

labels_dict={0:'MASK',1:'NO MASK'}
color_dict={0:(0,255,0),1:(0,0,255)}
```

```python
capture=cv2.VideoCapture(0)
data=[]
font=cv2.FONT_HERSHEY_COMPLEX
while(True):

    
    flag,img=capture.read()
    if flag:
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces=face_clsfr.detectMultiScale(gray,1.3,5)
        #faces=face_clsfr.detectMultiScale(img)

    for (x,y,w,h) in faces:
    
        face_img=gray[y:y+w,x:x+w]
        #face_img=img[y:y+w,x:x+w]
        resized=cv2.resize(face_img,(100,100))
        normalized=resized/255.0
        reshaped=np.reshape(normalized,(1,100,100,1))
        result=model.predict(reshaped)

        label=np.argmax(result,axis=1)[0]
      
        cv2.rectangle(img,(x,y),(x+w,y+h),color_dict[label],2)
        cv2.rectangle(img,(x,y-40),(x+w,y),color_dict[label],-1)
        cv2.putText(img, labels_dict[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
        
        
    cv2.imshow('LIVE',img)
    key=cv2.waitKey(1)
    
    if(key==27):
        break
        
cv2.destroyAllWindows()
capture.release()
```
